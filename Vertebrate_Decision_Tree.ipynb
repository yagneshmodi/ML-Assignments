```python
import pandas as pd
```

```python
data = pd.read_csv("/content/vertebrate.csv", header='infer')
```

```python
data
```

```python
data['Class'] = data['Class'].replace(['reptiles', 'fishes', 'amphibians', 'birds'], 'non-mammals')
```

```python
data
```

```python
pd.crosstab([data['Warm-blooded'],data['Gives Birth']],data['Class'])
```

```python
from sklearn import tree
```

```python
y = data['Class']
```

```python
y
```

```python
X = data.drop(['Name', 'Class'], axis=1)
```

```python
X
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
clf
```

```python
clf = clf.fit(X,y)
```

```python
import pydotplus
```

```python
from IPython.display import Image
```

```python
dot_data=tree.export_graphviz(clf, feature_names=X.columns,
                              class_names=['mammals', 'non-mammals'],
                              filled=True, out_file=None)
```

```python
dot_data
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
graph
Image(graph.create_png())
```

```python
testData = [['gila monister',0,0,0,0,1,1,'non-mammals'],
            ['platypus',1,0,0,0,1,1,'mammals'],
            ['owl', 1,0,0,1,1,0,'non-mammals'],
            ['dolphin',1,1,1,0,0,0,'mammals']]
```

```python
testData
```

```python
data.columns
```

```python
testData = pd.DataFrame(testData, columns=data.columns)
```

```python
testY = testData['Class']
```

```python
testY
```

```python
testX = testData.drop(['Name', 'Class'], axis=1)
```

```python
testX
```

```python
predY = clf.predict(testX)
```

```python
predY
```

```python
testY
```

```python
predictions = pd.concat([testData['Name'], testData['Class'], pd.Series(predY, name="Predicted Class")], axis=1)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(testY, predY)
```

```python
print("Accuracy on test data is %.2f" % (accuracy_score(testY, predY)*100))
```
